5. Data Selection – Based on Conditional filtering

```python
#Select observation with the value in rating column >=5
df_rating=df[df['Rating']>=5]
df_rating
```

8. Dealing with missing values

```python
df.isnull().sum()
```

```python
Rev_mean=df['Revenue (Millions)'].mean()
df['Revenue (Millions)'].fillna(Rev_mean)
```

```python
met_mean=df['Metascore'].mean()
df['Metascore'].fillna(met_mean)
```

```python
df.fillna(0)
```

9. Dropping columns and null values

```python
df.dropna(axis=1,how='all')
```
