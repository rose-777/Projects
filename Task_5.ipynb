5. Data Selection – Based on Conditional filtering

```python
#Select observation with the value in rating column >=5
df_rating=df[df['Rating']>=5]
df_rating
```

```python
#Select observation with the value in the revenue (Millions) column <100
df_revenue=df[df['Revenue (Millions)']<100]
df_revenue
```

Groupby operations

```python
#grouping by 'Genre'
df_genre=df.groupby('Genre')[['Director']].count()
df_genre
```

7. Sorting operation


```python
# sorting by 'Year'
df.sort_values('Year')
```

```python
#sorting by 'Rating'
df.sort_values('Rating',ascending=False)
```

```python
df_g=[df.Genre=='Action']
df_g
```

8. Dealing with missing values

```python
df.isnull().sum()
```

```python
Rev_mean=df['Revenue (Millions)'].mean()
df['Revenue (Millions)'].fillna(Rev_mean)
```

```python
met_mean=df['Metascore'].mean()
df['Metascore'].fillna(met_mean)
```

```python
df.fillna(0)
```

9. Dropping columns and null values

```python
df.dropna(axis=1,how='all')
```

10. Apply( ) functions

```python
df[['Revenue (Millions)','Rating']].agg(['min','mean','max'])
```

```python
df[['Votes','Metascore']].agg(['min','mean','max'])
```
